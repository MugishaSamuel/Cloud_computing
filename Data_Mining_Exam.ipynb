### b)	Determine the:
### i.	Mean absolute error. (3 marks)


```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error
y = df["MRI_Count"]
x = df[["Height"]]
lr = LinearRegression()
lr.fit(x,y)
Yhat = lr.predict(x)
print(f"The mean Absolute Error is {mean_absolute_error(y, Yhat)}")
```

### ii.	Residual sum of squares of the simple linear regression. (3 marks)

```python
RSS = ((Yhat - y) ** 2).sum()

print(f"Residual sum of squares of the simple linear regression is {RSS}")
```

### iii.	R-squared for the model. (3 marks)

```python
# The value of the R-squared is the percentage of variation of the response variable (y) that is explained by a linear model.
lr = LinearRegression()
lr.fit(x,y)
R = lr.score(x,y)

print(f"R-squared for the model is {(R*100).round(1)}%")
```
